Continuaremos con el pre-procesamiento de los datos, renombrando columnas con nombres mas manejables, eliminando filas y columnas duplicadas normalizando los tipos de datos de nuestras variables de interes y limpiando los titulos de las peliculas de nuestra base de datos.

Finalmente mostramos la base limpia y con nuestras columnas de interes.

```python
df1=df.rename(columns = {'Cast_Total_facebook_likes':'likesfb_elenco'})
df1=df1.rename(columns = {'movie_title':'Pelicula'})
df1=df1.rename(columns = {'title_year':'año'})
df1=df1.rename(columns = {'gross':'rec_bruto'})
df1=df1.rename(columns = {'DIRECTOR_facebook_likes':'likes_director'})
df1=df1.drop('title_year.1',axis = 1 )
y=[]
datos_x = df1["Pelicula"]
for i in datos_x:
    x=i.rstrip("?ÿ")
    y.append(x)
df1["Pelicula"]=y

director=df1["likes_director"]
director[4]='475'
df1["likes_director"]=director

data = df1[['año','Pelicula','rec_bruto','likesfb_elenco','likes_director','duration']].copy()
data1=data.drop(8,axis=0)
data1.head(14)

```

Podemos notar que nuestra base cuenta con valores nulos los cuales posteeriormente los identificaremos en nuestro analisis y los rellenaremos con ceros

```python
data1.isnull()
```

```python
data1= data1.fillna(0)
data1.head(13)
```

Consideraremos 3 diferentes agrupaciones:

1. Para identidficar cuales son las peliculas con recaudacion bruta encima del promedio
2. cuales son las peliculas mas largas de 2 horas
3. peliculas posteriores al 2008

```python
x=data1['rec_bruto']
n=len(x)
s=0
for i in x:
    s=s+i
p=s/n


clase_1 = data1[data1['rec_bruto']>p]
clase_1 =clase_1.drop(['año','likesfb_elenco','likes_director','duration'],axis = 1 )
clase_1.to_csv('clase1.csv', index=False)

clase_2 = data1[data1['duration']>120]
clase_2 =clase_2.drop(['año','likesfb_elenco','likes_director'],axis = 1 )
clase_2.to_csv('clase2.csv', index=False)

clase_3 = data1[data1['año']>2008]
clase_3 =clase_3.drop(['año','likesfb_elenco','likes_director'],axis = 1 )
clase_3.to_csv('clase3.csv', index=False)
```

Con las tres distintas clases, creamos nuevos archivos

```python
clase1 = pd.read_csv("clase1.csv",encoding='utf8')
clase2 = pd.read_csv("clase2.csv",encoding='utf8')
clase3 = pd.read_csv("clase3.csv",encoding='utf8')
```
