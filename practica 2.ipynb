#### Visualizacion de los datos

```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("winemag-data_first150k.csv",encoding='utf8')
df.head(10)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df=df.rename(columns = {'Unnamed: 0':'clave'})
```

```python
df = df[['clave','country','points','price','province','variety']].copy()
df.isnull()
df= df.fillna(0)
df
```

```python
x=df['points'].mean()
y=df['price'].mean()
print('Promedio de puntos y precios')
print(x,y)
df.corr()
```

De acuerdo con la siguiente grafica mas vinos se encuentran entre el promedio de precios y puntos. se representa en la grafica con los puntos mas obscuros

```python
df[df['price'] < 95].plot.hexbin(x='price', y='points', gridsize=18)
```

En esta parte analizamos los Vinos de chile y el Historgrama nos indica que hay mas frecencia en vinos con un pntaje de alrededor de 86

```python
df1 = df[df['country']=='Chile']
df1[df1['points'] < 100]['points'].plot.hist()
df1['points'].describe()
```

###### Correlacion entre precio y puntaje vinos chilenos

```python
sns.jointplot(x='price', y='points', data=df1[df1['price'] < 100])
```

###### Paises con vinos mejor puntuados

```python
df2 = df[df['points']>88]
df2['country'].value_counts().head(13).plot.bar()
```

Con esta grafica podemos denotar que US tiene mas vinos con puntuacion mayor a 88, es decir que tine los mejores vinos

###### Francia vs Italia correlacion 

```python
df3 = df[df['country']=='France']
df4 = df[df['country']=='Italy']
df3 = pd.concat([df3,df4])
sns.lmplot(x='points', y='price', hue='country', 
           data=df3.loc[df3['country'].isin(['France', 'Italy'])], 
           fit_reg=False)
```
